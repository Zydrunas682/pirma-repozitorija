```python
import requests
from bs4 import BeautifulSoup

# Siųsti GET užklausą į svetainę
url = "https://www.skelbimai.lt/skelbimai/nekilnojamasis-turtas/butai"
response = requests.get(url)

# Patikrinti, ar užklausa buvo sėkminga
if response.status_code == 200:
    # Apdoroti HTML turinį su BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Rasti visus img elementus (nuotraukos)
    img_tags = soup.find_all('img')  # Paieška visiems img tagams

    if img_tags:
        for i, img_tag in enumerate(img_tags, start=1):
            # Ištraukti nuotraukos adresą (src atributas)
            img_url = img_tag.get('src')
            
            if img_url:
                print(f"Nuotraukos {i} adresas: {img_url}")
            else:
                print(f"Nuotrauka {i} neturi URL.")
    else:
        print("Nuotraukos nerastos.")
else:
    print(f"Įvyko klaida, kodas: {response.status_code}")
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
opcijos = Options()
opcijos.add_argument('--incognito')
opcijos.headless = True  
driver = webdriver.Chrome(options=opcijos)
base_url = "https://www.baldoras.lt/katalogas/valgomojo-stalai/"
driver.get(base_url)
prices = []
    
page = 1
while True:
    print(f"Scraping page {page}...")
    
    source = driver.page_source
    soup = BeautifulSoup(source, "html.parser")
    
    price_elements = soup.find_all("div", class_="product-card__actions")
    
    for price_element in price_elements:
        price = price_element.find("div", class_="product-card__prices")
        if price:
            try:
                price_value = price.get_text(strip=True).replace("€", "").replace(",", ".")
                price_value = float(price_value)
                prices.append(price_value)
                print(price_value)
            except ValueError:
                pass  
    
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "li.page-item a[aria-label='Next']")  
        next_button.click()
        WebDriverWait(driver, 10).until(EC.staleness_of(next_button))  
        sleep(2) 
        page += 1  
    except Exception as e:
        print(f"No next page found or error")
        break  

if prices:
    average_price = sum(prices) / len(prices)
    print(f"Average price: €{average_price:.2f}")
else:
    print("No prices found.")
```
