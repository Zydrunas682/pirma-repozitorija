```python
# Sukurkite Mokinys klasę, kuri turės mokinio vardą, pavardę ir sąrašą pažymių. Įtraukite metodus, kurie skaičiuotų vidurkį, didžiausią ir mažiausią pažymį. 



class Mokinys:
    def __init__(self, vardas, pavarde, ivertinimai):
        # Konstruktorius, kuris nustato mokinio vardą, pavardę ir pažymių sąrašą
        self.vardas = vardas
        self.pavarde = pavarde
        self.ivertinimai = ivertinimai

    def vidurkis(self):
        """Apskaičiuoja ir grąžina pažymių vidurkį."""
        if len(self.ivertinimai) == 0:  # Jei nėra pažymių
            return 0  # Grąžiname 0, jei sąrašas tuščias
        return sum(self.ivertinimai) / len(self.ivertinimai)  # Teisingas vidurkio skaičiavimas

    def didziausias_ivertinimas(self):
        """Grąžina didžiausią įvertinimą."""
        if len(self.ivertinimai) == 0:  # Jei nėra pažymių
            return None  # Grąžiname None, jei sąrašas tuščias
        return max(self.ivertinimai)  # Grąžiname didžiausią įvertinimą

    def maziausias_ivertinimas(self):
        """Grąžina mažiausią įvertinimą."""
        if len(self.ivertinimai) == 0:  # Jei nėra pažymių
            return None  # Grąžiname None, jei sąrašas tuščias
        return min(self.ivertinimai)  # Grąžiname mažiausią įvertinimą

    def __str__(self):
        """Grąžina mokinio informaciją kaip tekstą."""
        # Išvedame vardą, pavardę, vidurkį, didžiausią ir mažiausią pažymį
        return f"{self.vardas} {self.pavarde}, Vidurkis: {self.vidurkis():.2f}, Didžiausias įvertinimas: {self.didziausias_ivertinimas()}, Mažiausias įvertinimas: {self.maziausias_ivertinimas()}"


# Sukuriame mokinius su vardu, pavarde ir pažymių sąrašais
mokinys1 = Mokinys("Jonas", "Jonaitis", [7, 8, 9, 10])
mokinys2 = Mokinys("Zydrunas", "Andriulis", [3, 9, 9, 7])
mokinys3 = Mokinys("Diana","Zuikis", [2, 4, 9, 6])

# Išvedame mokinių informaciją
print(mokinys1)  # Išveda mokinio 1 informaciją
print(mokinys2)  # Išveda mokinio 2 informaciją
print(mokinys3)


# Apskaičiuojame ir išvedame mokinių vidurkius, didžiausius ir mažiausius įvertinimus
print(f"Vidurkis (Jonas): {mokinys1.vidurkis()}")  # Vidurkis pirmam mokiniui
print(f"Vidurkis (Zydrunas): {mokinys2.vidurkis()}")  # Vidurkis antram mokiniui
print(f"vidurkis (Diana): {mokinys3.vidurkis()}")

print(f"Didžiausias įvertinimas (Jonas): {mokinys1.didziausias_ivertinimas()}")
print(f"Didžiausias įvertinimas (Zydrunas): {mokinys2.didziausias_ivertinimas()}")
print(f"Didžiausias įvertinimas (Diana): {mokinys3.didziausias_ivertinimas()}")


print(f"Mažiausias įvertinimas (Jonas): {mokinys1.maziausias_ivertinimas()}")
print(f"Mažiausias įvertinimas (Zydrunas): {mokinys2.maziausias_ivertinimas()}")
print(f"Mažiausias įvertinimas (Diana): {mokinys3.maziausias_ivertinimas()}")


class Abiturientas(Mokinys):
    def __init__(self, vardas, pavarde, ivertinimai, egzamino_rezultatas=None):
        # Naudojame tėvų klasės konstruktorių (Mokinys)
        super().__init__(vardas, pavarde, ivertinimai)
        self.egzamino_rezultatas = egzamino_rezultatas  # Egzamino rezultatas

    def prideti_egzamina(self, rezultatas):
        """Prideda egzamino rezultatą."""
        self.egzamino_rezultatas = rezultatas

    def bendras_vidurkis(self):
        """Apskaičiuoja bendrą vidurkį, įtraukiant pažymius ir egzamino rezultatą."""
        if self.egzamino_rezultatas is not None:  # Jei egzaminas yra
            return (sum(self.ivertinimai) + self.egzamino_rezultatas) / (len(self.ivertinimai) + 1)
        else:  # Jei egzaminas nėra
            return self.vidurkis()

    def __str__(self):
        """Grąžina abituriento informaciją kaip tekstą."""
        if self.egzamino_rezultatas is not None:
            return f"{self.vardas} {self.pavarde}, Vidurkis (įskaitant egzaminą): {self.bendras_vidurkis():.2f}, Egzamino rezultatas: {self.egzamino_rezultatas}, Didžiausias įvertinimas: {self.didziausias_ivertinimas()}, Mažiausias įvertinimas: {self.maziausias_ivertinimas()}"
        else:
            return f"{self.vardas} {self.pavarde}, Vidurkis: {self.vidurkis():.2f}, Didžiausias įvertinimas: {self.didziausias_ivertinimas()}, Mažiausias įvertinimas: {self.maziausias_ivertinimas()}"

# Sukuriame abiturientą su vardu, pavarde, pažymiais ir egzamino rezultatu
abiturientas1 = Abiturientas("Jonas", "Jonaitis", [7, 8, 9, 10], 9)
abiturientas2 = Abiturientas("Zydrunas", "Andriulis", [3, 9, 9, 7])

# Išvedame abiturientų informaciją
print(abiturientas1)  # Išveda abituriento 1 informaciją
print(abiturientas2)  # Išveda abituriento 2 informaciją

# Apskaičiuojame ir išvedame abiturientų bendrą vidurkį
print(f"Bendras vidurkis (Jonas): {abiturientas1.bendras_vidurkis()}")  # Bendras vidurkis
print(f"Bendras vidurkis (Zydrunas): {abiturientas2.bendras_vidurkis()}")  # Bendras vid
```

```python
# 3  Sukurkite Mokykla klasę, kuri turės sąrašą Mokinys objektų.
#  Įtraukite metodus, kurie leistų pridėti naują mokinį,
# 
#  pašalinti mokinį, bei skaičiuoti visos mokyklos mokinių pažymių 
# vidurkius.

class Mokinys:
    def __init__(self, vardas, pavarde, ivertinimai):
        # Konstruktorius, kuris nustato mokinio vardą, pavardę ir pažymių sąrašą
        self.vardas = vardas
        self.pavarde = pavarde
        self.ivertinimai = ivertinimai

    def vidurkis(self):
        """Apskaičiuoja ir grąžina pažymių vidurkį."""
        if len(self.ivertinimai) == 0:
            return 0  # Jei nėra pažymių
        return sum(self.ivertinimai) / len(self.ivertinimai)

    def __str__(self):
        """Grąžina mokinio informaciją kaip tekstą."""
        return f"{self.vardas} {self.pavarde}, Vidurkis: {self.vidurkis():.2f}"

class Mokykla:
    def __init__(self, pavadinimas):
        # Konstruktorius, kuris nustato mokyklos pavadinimą ir tuščią mokinių sąrašą
        self.pavadinimas = pavadinimas
        self.mokiniai = []

    def prideti_mokini(self, mokinys):
        """Prideda naują mokinį į mokyklą."""
        self.mokiniai.append(mokinys)

    def pasalinti_mokini(self, vardas, pavarde):
        """Pašalina mokinį iš mokyklos pagal vardą ir pavardę."""
        for mokinys in self.mokiniai:
            if mokinys.vardas == vardas and mokinys.pavarde == pavarde:
                self.mokiniai.remove(mokinys)
                print(f"Mokinys {vardas} {pavarde} buvo pašalintas iš mokyklos.")
                return
        print(f"Mokinys {vardas} {pavarde} nerastas mokykloje.")

    def vidurkis(self):
        """Apskaičiuoja visų mokinių pažymių vidurkį mokykloje."""
        if len(self.mokiniai) == 0:
            return 0  # Jei nėra mokinių
        bendras_vidurkis = sum(mokinys.vidurkis() for mokinys in self.mokiniai) / len(self.mokiniai)
        return bendras_vidurkis

    def __str__(self):
        """Grąžina mokyklos informaciją su visų mokinių sąrašu."""
        mokiniu_sarasas = "\n".join(str(mokinys) for mokinys in self.mokiniai)
        return f"Mokykla: {self.pavadinimas}\nMokiniai:\n{mokiniu_sarasas}"

# Sukuriame mokyklą
mokykla = Mokykla("Vilniaus Gimnazija")

# Sukuriame mokinius
mokinys1 = Mokinys("Jonas", "Jonaitis", [7, 8, 9, 10])
mokinys2 = Mokinys("Zydrunas", "Andriulis", [3, 9, 9, 7])
mokinys3 = Mokinys("Eglė", "Povilaitytė", [8, 8, 9, 9])

# Pridedame mokinius į mokyklą
mokykla.prideti_mokini(mokinys1)
mokykla.prideti_mokini(mokinys2)
mokykla.prideti_mokini(mokinys3)

# Išvedame visą mokyklos informaciją
print(mokykla)

# Apskaičiuojame ir išvedame mokyklos vidurkį
print(f"Mokyklos vidurkis: {mokykla.vidurkis():.2f}")

# Pašaliname mokinį
mokykla.pasalinti_mokini("Zydrunas", "Andriulis")

# Išvedame naują mokyklos informaciją
print("\nPo pašalinimo:")
print(mokykla)

# Apskaičiuojame ir išvedame naują mokyklos vidurkį po pašalinimo
print(f"Mokyklos vidurkis po pašalinimo: {mokykla.vidurkis():.2f}")


```
